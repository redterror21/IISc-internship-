# Example 1 = Steady State Heat Equation
In this eg we solve the following sse
\begin{equation}
\frac{\mathrm{d}^2 T}{\mathrm{d}x^2} = H(T-T_\infty)
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1.0 #lenght of rod
n= 101
x = np.linspace(0,L,n) #to get discrete x axis values
dx = x[1] - x[0]
H = 0.2
Tleft = 200
Tright = 300
Tinf = 500

print(dx)
```

```python
#build the diagonals
ud = np.ones(n-1)
ud[0]=0
ld = np.ones(n-1)
ld[-1]=0

#main
md = -( 2+H*dx*dx)*np.ones(n)
md[0]=1
md[-1]=1

#full matrix
A = np.diag(ld,-1) + np.diag(ud,1) + np.diag(md,0)
print(A)

rhs = -H*dx*dx*Tinf*np.ones(n)
rhs[0] = Tleft
rhs[-1] = Tright
print(rhs)
```

```python
sol=np.linalg.solve(A,rhs)
print(sol)
```

```python
plt.plot(x,sol)
```
